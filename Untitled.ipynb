```python
def label_race (row):
    MonthlyCharges = row["MonthlyCharges"]
    res = 0
    
    p_25 = 38.5
    p_50 = 70.65
    p_75 = 90.45

    if (MonthlyCharges < p_25):
      res = 1
    elif ((MonthlyCharges >= p_25) and (MonthlyCharges < p_50)):
      res = 2
    elif ((MonthlyCharges >= p_50) and (MonthlyCharges < p_75)):
      res = 3
    else:
      res = 4
    return res

df_train["MonthlyCharges_type"] = df_train.apply(lambda row: label_race(row), axis=1)
df_test["MonthlyCharges_type"] = df_test.apply(lambda row: label_race(row), axis=1)
```

```python
col = "MonthlyCharges_type"
df_train_temp = pd.get_dummies(df_train[col], prefix=col)
df_train = df_train.join(df_train_temp)
df_test_temp = pd.get_dummies(df_test[col], prefix=col)
df_test = df_test.join(df_test_temp)
```

```python
df_train = df_train.drop([col], axis = 1)
df_test = df_test.drop([col], axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_train["Tenure"] = scaler.fit_transform(df_train[["Tenure"]])
df_test["Tenure"] = scaler.transform(df_test[["Tenure"]])

df_train["TotalCharges"] = scaler.fit_transform(df_train[["TotalCharges"]])
df_test["TotalCharges"] = scaler.transform(df_test[["TotalCharges"]])

df_train["MonthlyCharges"] = scaler.fit_transform(df_train[["MonthlyCharges"]])
df_test["MonthlyCharges"] = scaler.transform(df_test[["MonthlyCharges"]])
```

```python
false_df = df_train[df_train["Churn"] == 0].sample(frac=1)
true_df = df_train[df_train["Churn"] == 1].sample(frac=1)

df_train_true_scope_1 = true_df.iloc[:952]
df_train_false_scope_1 = false_df.iloc[:952]
df_all_train_scope_1 = df_train_true_scope_1.append(df_train_false_scope_1).sample(frac=1)


df_train_true_scope_2 = true_df.iloc[:952]
df_train_false_scope_2 = false_df.iloc[952:1904]
df_all_train_scope_2 = df_train_true_scope_2.append(df_train_false_scope_2).sample(frac=1)


df_train_true_scope_3 = true_df.iloc[:735]
df_train_false_scope_3 = false_df.iloc[1904:2639]
df_all_train_scope_3 = df_train_true_scope_3.append(df_train_false_scope_3).sample(frac=1)
```

```python
from sklearn.metrics import matthews_corrcoef
simple_space_xgb = {
            'n_estimators': hp.choice('n_estimators', range(100, 1000)),
            'eta': hp.quniform('eta', 0.025, 0.5, 0.025),
            'max_depth':  hp.choice('max_depth', np.arange(1, 4, dtype=int)),
            'objective': "binary:logistic",
}
```

```python
X1 = df_all_train_scope_1.drop(["Churn"], axis=1)
y1 = df_all_train_scope_1["Churn"]

X_train1, X_valid1, y_train1, y_valid1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

def hyperopt_xgb_score(params):
    clf = xgboost.XGBClassifier(**params)
    current_score = cross_val_score(clf, X_train1, y_train1, cv=5, scoring="f1").mean()
    print("f1 train:")
    print(current_score, params)

    clf.fit(X_train1, y_train1)
    y_pred = clf.predict(X_valid1)
    score_2 = matthews_corrcoef(y_valid1, y_pred)
    print("matthews_corrcoef valid:")
    print(score_2)

    print()
    print("----------------------------")
    print()
    return -score_2

best1 = fmin(fn=hyperopt_xgb_score, space=simple_space_xgb, algo=tpe.suggest, max_evals=20)
print(best1)
```

```python
best_params_1 = {'objective': 'binary:logistic', 
        'max_depth': best1['max_depth'],
        'eta': best1['eta'],
        'n_estimators': best1['n_estimators']}
```

```python
X2 = df_all_train_scope_2.drop(["Churn"], axis=1)
y2 = df_all_train_scope_2["Churn"]

X_train2, X_valid2, y_train2, y_valid2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

def hyperopt_xgb_score(params):
    clf = xgboost.XGBClassifier(**params)
    current_score = cross_val_score(clf, X_train2, y_train2, cv=5, scoring="f1").mean()
    print("f1 train:")
    print(current_score, params)

    clf.fit(X_train2, y_train2)
    y_pred = clf.predict(X_valid2)
    score_2 = matthews_corrcoef(y_valid2, y_pred)
    print("matthews_corrcoef valid:")
    print(score_2)

    print()
    print("----------------------------")
    print()
    return -score_2

best2 = fmin(fn=hyperopt_xgb_score, space=simple_space_xgb, algo=tpe.suggest, max_evals=20)
print(best2)
```

```python
best_params_2 = {'objective': 'binary:logistic', 
        'max_depth': best2['max_depth'],
        'eta': best2['eta'],
        'n_estimators': best2['n_estimators']}
```

```python
X3 = df_all_train_scope_3.drop(["Churn"], axis=1)
y3 = df_all_train_scope_3["Churn"]

X_train3, X_valid3, y_train3, y_valid3 = train_test_split(X3, y3, test_size=0.2, random_state=42)

def hyperopt_xgb_score(params):
    clf = xgboost.XGBClassifier(**params)
    current_score = cross_val_score(clf, X_train3, y_train3, cv=5, scoring="f1").mean()
    print("f1 train:")
    print(current_score, params)

    clf.fit(X_train3, y_train3)
    y_pred = clf.predict(X_valid3)
    score_2 = matthews_corrcoef(y_valid3, y_pred)
    print("matthews_corrcoef valid:")
    print(score_2)

    print()
    print("----------------------------")
    print()
    return -score_2

best3 = fmin(fn=hyperopt_xgb_score, space=simple_space_xgb, algo=tpe.suggest, max_evals=20)
print(best3)
```

```python
best_params_3 = {'objective': 'binary:logistic', 
        'max_depth': best3['max_depth'],
        'eta': best3['eta'],
        'n_estimators': best3['n_estimators']}
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
xgb_model1 = XGBClassifier(**best_params_1)
xgb_model2 = XGBClassifier(**best_params_2)
xgb_model3 = XGBClassifier(**best_params_3)

xgb_model1.fit(train_x, train_y)
predict = xgb_model1.predict(train_x)
predict_arr = np.array(predict).round()
print(matthews_corrcoef(train_y, predict_arr))

xgb_model2.fit(train_x, train_y)
predict = xgb_model2.predict(train_x)
predict_arr = np.array(predict).round()
print(matthews_corrcoef(train_y, predict_arr))

xgb_model3.fit(train_x, train_y)
predict = xgb_model3.predict(train_x)
predict_arr = np.array(predict).round()
print(matthews_corrcoef(train_y, predict_arr))

predict = (xgb_model2.predict(train_x) + xgb_model3.predict(train_x)) / 2
predict_arr = np.array(predict).round()
print(matthews_corrcoef(train_y, predict_arr))
```
